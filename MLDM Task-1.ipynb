```python
# Import necessary libraries
import pandas as pd

# Load the dataset
df = pd.read_csv('online_shoppers_intention 2.csv')

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Get data types and unique values of columns
print("Data types and unique values:\n", df.dtypes)

# Review target variable distribution
print("Target distribution:\n", df['Revenue'].value_counts())
```

```python
# Fill missing values if any (you may adjust based on the specific dataset requirements)
df.fillna(method='ffill', inplace=True)

# Encode categorical variables
df = pd.get_dummies(df, drop_first=True)
```

```python
# Scale numerical features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_features] = scaler.fit_transform(df[numerical_features])

# Define features and target
X = df.drop('Revenue', axis=1)  # Replace 'Revenue' with your target column name if different
y = df['Revenue']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Initialize KNN with k=5 and fit
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
# Predict and evaluate
y_pred_knn = knn.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
print("KNN Classification Report:\n", classification_report(y_test, y_pred_knn))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Initialize Decision Tree classifier and fit
dt = DecisionTreeClassifier(max_depth=5, criterion='gini', random_state=42)
dt.fit(X_train, y_train)
```

```python
# Predict and evaluate
y_pred_dt = dt.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Decision Tree Classification Report:\n", classification_report(y_test, y_pred_dt))
```
